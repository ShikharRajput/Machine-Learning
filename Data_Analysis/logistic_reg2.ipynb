```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
file = pd.read_csv("Dataset-master\Social_Network_Ads.csv")
```

```python
file.tail()
```

```python
file.corr()
```

```python
x,y = file[['Age','EstimatedSalary']], file['Purchased']
x.head(),y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.99, random_state=2)
X=x_test
```

```python
x_train.tail(),x_test.tail(),y_train.tail(),y_test.tail()
```

```python
lr = LinearRegression()
lr1 = lr.fit(x_train,y_train)
```

```python
pre = lr.predict(x_test)
pre[-1:-10:-1]
```
