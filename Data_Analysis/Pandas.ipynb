```python
# Return a fixed frequency DatetimeIndex, with day (calendar) as the default frequency
date1 = pd.date_range(start = "29-09-1999", periods = 19*12, freq = "M")
print(date1)
```

```python
time = pd.date_range(start = "28-06-2019", periods = 25, freq = "H")
time
```

```python
# tz : string, default None Time zone name for returning localized DatetimeIndex, for example Asia/Hong_Kong
time1 = pd.date_range(start = "28-06-2019", periods = 25, freq = "H", tz = "Asia/Hong_Kong")
time1
```

```python
# normalize : bool, default False Normalize start/end dates to midnight before generating date range
time2 = pd.date_range(start = "12:15:30", periods = 25, freq = "H", normalize = False)
time2
```

```python
# steps in hours and slicing
time3 = pd.date_range(start = "12:15:30", periods = 25, freq = "3H", normalize = True)
print(time3)
time3[0:10:2]
```

```python
# The time machine and slicing
time3 = pd.date_range(start = "12:15:30", periods = 25, freq = "-12H", normalize = False)
print(time3)
time3[0:10:2]
```
